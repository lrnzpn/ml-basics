### Plotting in Python using matplotlib.pyplot

````python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

````python
# For compatibility across multiple platforms
import os
IB = os.environ.get('INSTABASE_URI',None) is not None
open = ib.open if IB else open
````

````python
# Read Cities.csv and Countries.csv into lists of dictionaries
cities = []
with open('Cities.csv','r') as f:
    rows = csv.DictReader(f)
    for r in rows:
        cities.append(r)
countries = []
with open('Countries.csv','r') as f:
    rows = csv.DictReader(f)
    for r in rows:
        countries.append(r)
````

````python
# Introduction to scatterplots
x = [1, 2, 4, 6, 9]
y = [4, 5, 2, 7, 5]
plt.scatter(x, y)
plt.show()
````

````python
# Scatterplot of temperature (y) versus latitude (x)
x = []
y = []
for city in cities:
    x.append(city['latitude'])
    y.append(city['temperature'])
plt.xlabel('latitude')
plt.ylabel('temperature')
plt.scatter(x, y)
plt.show()
````

````python
# Scatterplot of latitude (y) versus longitude (x) colored by temperature
x = []
y = []
colors = []
for city in cities:
    x.append(city['longitude'])
    y.append(city['latitude'])
    if float(city['temperature']) < 8:
        colors.append('blue')
    elif float(city['temperature']) < 12:
        colors.append('red')
    else: colors.append('green')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.scatter(x, y, c=colors)
plt.show()
x
````

````python
# Introduction to bar charts
bars = ['Bar1', 'Bar2', 'Bar3','Bar4']
heights = [2, 6, 4,8]
plt.xticks([1,2,3,4], bars)
plt.bar([1,2,3,4], heights, align='center')
plt.show()
````

````python
# Bar chart showing populations of countries with 'ia' in their name
bars = [] # countries with 'ia' in their name
heights = [] # population of those countries
for country in countries:
    if 'ia' in country['country']:
        bars.append(country['country'])
        heights.append(country['population'])
numbars = len(bars)
# range(numbars) in following command yields [0,1,...,numbars-1]
plt.xticks(range(numbars), bars, rotation='vertical')
plt.bar(range(numbars), heights, align='center')
plt.xlabel('country')
plt.ylabel('population')
plt.show()
````

````python
# Introduction to pie charts
sizes = [10, 20, 30]
lbls = ['small', 'medium', 'large']
plt.pie(sizes, labels=lbls, autopct='%1.1f%%')
plt.show()
````

````python
# Pie chart showing number of EU countries versus non-EU countries
numEU = 0
numNotEU = 0
for country in countries:
    if country['EU'] == 'yes':
        numEU += 1
    else:
        numNotEU +=1
sizes = [numEU, numNotEU]
lbls = ['EU', 'not EU']
plt.pie(sizes, labels=lbls, autopct='%1.1f%%')
plt.show()
````

#### Read ```Players.csv``` and ```Teams.csv``` into lists of dictionaries

````python
# Read Players.csv and Teams.csv into lists of dictionaries
players = []
with open('Players.csv','r') as f:
    rows = csv.DictReader(f)
    for r in rows:
        players.append(r)
teams = []
with open('Teams.csv','r') as f:
    rows = csv.DictReader(f)
    for r in rows:
        teams.append(r)
````

### Create a scatterplot of players showing ```passes made (y-axis) versus minutes played (x-axis)```. 
<b>Color each player based on their position (goalkeeper, defender, midfielder, forward)</b>.
<br>Hint: remember copy-paste-modify approach to programming!

````python
# your code here
from matplotlib.lines import Line2D

x = []
y = []
c = []
positions = ['midfielder', 'defender', 'goalkeeper', 'forward']

for player in players:
    x.append(int(player['minutes']))
    y.append(int(player['passes']))
    if player['position'] == positions[0]:
        c.append('yellow')
    elif player['position'] == positions[1]:
        c.append('red')
    elif player['position'] == positions[2]:
        c.append('blue')
    else: 
        c.append('green')


colors = ['yellow','red','blue','green']
dot = []
for i in range(len(colors)):
    # needed a workaround to make the legend display without changing much
    dot.append(Line2D([0], [0], marker='o', color='w', label=positions[i], markerfacecolor=colors[i], markersize=8))
plt.legend(dot, positions, loc=2)

plt.scatter(x,y,c=c)
plt.title("Scatterplot of all players\' passes made vs minutes played")
plt.xlabel('Minutes Played')
plt.ylabel('Passes Made')  
plt.show()
````

Create a simple <b>LOOP</b> program that asks for a team name then plots a pie chart of the percentages of the positions per team(midfielder,defender,goalkeeper,forward), the loop must break when the user enters <i>quit</i>, If the inputted country is not available, the program must output the string ```Team Not Available in the World Cup.```

A sample prompt is given at the code output below. 



````python
# your code here
while True:
    team_name = input('Enter Team Name:')
    if team_name == 'quit':
        print('Team not available in world cup')
        break
    # check if team is in the list of teams that participated
    elif team_name not in [team['\ufeffteam'] for team in teams]: 
        print('Team not available in world cup')
    else:
        mf_count = 0
        fw_count = 0
        gp_count = 0
        df_count = 0
        for player in players:
            if team_name == player['team']:
                if player['position'] == 'midfielder':
                    mf_count += 1
                elif player['position'] == 'defender':
                    df_count += 1
                elif player['position'] == 'forward':
                    fw_count += 1
                else:
                    gp_count += 1
                    
        team_count = [mf_count, df_count, gp_count, fw_count]
        labels = ['midfielder', 'defender','goalkeeper','forward']
        plt.pie(team_count,labels=labels,autopct='%1.1f%%')
        plt.show()
````
